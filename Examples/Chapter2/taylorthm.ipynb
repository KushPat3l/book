```python
import numpy as np
import matplotlib.pyplot as plt

def Nth_term(x,N):
    return -(-1)**N*x**N/N
    
x = 0.4
N_max = 15 
Nlist = np.arange(1,N_max+1)
P = 0
PNlist = []
lowlist=[]
hilist=[]

for N in Nlist:
    P = P + Nth_term(x,N)
    PNlist.append(P)
    Np= N+1
    low = (x/(1+x))**Np/Np
    lowlist.append(low)    
    hi = x**Np/Np
    hilist.append(hi)   
    
RN = abs(PNlist-np.log(1+x))
        
plt.semilogy(Nlist, RN, lw=2, 
             color = 'r')
plt.semilogy(Nlist,lowlist,'r:',
             Nlist,hilist, 'r:')
plt.legend([r'$|R_N(x=0.4)|$']) 
plt.xticks(Nlist) 
plt.xlim([1,N_max])
plt.xlabel('Polynomial order N')
plt.ylim([1e-10,0.1])
plt.grid('on')
#plt.savefig('figTTlog.pdf')
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def Pksin(x,k):
    if (k%2) == 0:
        return 0
    else: 
        return (-1)**((k-1)/2)*x**k\
        /np.math.factorial(k)
    
# max degree of polynomial
N_max = 15 
x = 0.4
Psin = 0
Pksinlist = []
klist = np.arange(1,N_max+1)
lowsinlist=[]
upsinlist=[]

for k in klist:
    Psin = Psin + Pksin(x,k)
    Pksinlist.append(Psin)
    if (k%2)==0:
        upsin =x**(k+1)/np.math.factorial(k+1)
        lowsin = abs(np.cos(x))*upsin
    else:
        upsin =x**(k+2)/np.math.factorial(k+2)
        lowsin = abs(np.cos(x))*upsin
    kp = k+1
    lowsinlist.append(lowsin)    
    upsinlist.append(upsin)   
    
#Actual = np.full_like(Pklist, np.sin(0.5))
Eksin = abs(Pksinlist-np.sin(x))
        
plt.semilogy(klist,Eksin, lw=1, color='k')
plt.semilogy(klist, lowsinlist, 'r:', klist, upsinlist, 'r:')
plt.legend([r'$|R_N(x=0.4)|$']) 
plt.xticks(klist) 
plt.xlim([1,N_max])
plt.xlabel('Polynomial order N')
plt.ylim([1e-17,1e0])
plt.grid('on')
#plt.show()
plt.savefig('figTTsin.pdf')
```
