```python
import numpy as np
import matplotlib.pyplot as plt

def nth_term(x,n):
        return (-1)**n*x**(2*n+1)/\
        np.math.factorial(2*n+1)
    
S = 0
n_max = 6
x = np.linspace(0, 2*np.pi, 100)

for n in np.arange(0,n_max+1):
    S += nth_term(x,n)
    plt.plot(x,S,label='_nolegend_')
    
plt.plot(x, np.sin(x), lw=2, color='k')
plt.legend(['y=sin(x)'], 
           loc = 'lower left')
plt.xlim([0, 2*np.pi])
plt.ylim([-2,2])
plt.grid('on')
plt.show()
#plt.savefig('figsin.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def nth_term(x,n):
        return -(-1)**n*x**n/n
    
S = 0
n_max = 40
x = np.linspace(-0.99, 2, 100)

for n in np.arange(1,n_max+1):
    S += nth_term(x,n)
    b=n/n_max
    if (n>=5):
        plt.plot(x,S,label='_nolegend_',
        color = (0,0,b), lw=1)
    
plt.plot(x, np.log(1+x),lw=2,color='k')
plt.legend(['y=ln(1+x)'], 
           loc = 'upper left')
plt.xlim([-0.99, 2])
plt.ylim([-2,2])
plt.grid('on')
plt.show()
#plt.savefig('figlog.pdf')
```
