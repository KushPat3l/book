```python
import numpy as np
import matplotlib.pyplot as plt

def f0(x):
    return abs(np.round(x)-x)

m_max = 10
x = np.linspace(0,1,1000)

plt.plot(x,f0(x))

for i in np.arange(1,m_max+1):
    plt.plot(x,f0(2**i*x)/2**i)
    
plt.xlim([0,1])
plt.ylim([0,0.5])
plt.legend(['f0(x)','f1(x)',
            'f2(x)','f3(x)'])
plt.grid('on')
plt.show()
#plt.savefig('figsawtooth1.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f0(x):
    return abs(np.round(x)-x)

m_max = 10
x = np.linspace(0,1,1000)

def g(x):
    S = 0
    for i in np.arange(0,m_max+1):
        S += f0(2**i*x)/2**i
    return S

plt.xlim([0,1])
plt.ylim([0,0.7])
plt.plot(x, g(x),'k')
plt.legend(['g(x)'])
plt.grid('on')
#plt.show()
plt.savefig('figblancmange.pdf')
```
