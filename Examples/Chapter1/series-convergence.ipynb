```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(1,11)

A=np.zeros(len(x))
B=np.zeros(len(x))
C=np.zeros(len(x))
D=np.zeros(len(x))

def afunc(k):
    return 1/k**2

def bfunc(k):
    return (-1)**(k+1)/k

def cfunc(k):
    return 1/np.sqrt(k)

def dfunc(k):
    return 1/(1+np.log(k))

A[0] = afunc(1)
B[0] = bfunc(1)
C[0] = cfunc(1)
D[0] = dfunc(1)


for i in range(1,len(x)):
    A[i] = A[i-1]+afunc(x[i])
    B[i] = B[i-1]+bfunc(x[i])
    C[i] = C[i-1]+cfunc(x[i])
    D[i] = D[i-1]+dfunc(x[i])


plt.plot(x,A,x, B, '--', x, C, '-.', x, D, ':')
plt.xlabel('Number of terms')
plt.ylabel('Partial sums')
plt.legend(['a', 'b', 'c', 'd'])
plt.grid('on')
plt.savefig('figconv.pdf')
```
