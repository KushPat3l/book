```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(np.pi*x)

h=1e-6
x= np.linspace(-1,1,200)
fx= f(x)
fx1 = f(x+h)
df = (fx1-fx)/h

plt.plot(x,fx, '--', x, df)
plt.legend(["y=f(x)","y=f'(x)"])
plt.grid('on')
plt.xlim([-1,1])
plt.ylim([-3.2, 3.2])
plt.show()
            
```

```python
def g(x):
    return np.sqrt(np.abs(x))

h=1e-6
xp= np.linspace(0,1)
xn= np.linspace(-1,-h)
gxp = g(xp)
gxn = g(xn)
dgp = (g(xp+h) - gxp)/h
dgn = (g(xn+h) - gxn)/h

plt.plot(xp, gxp, 'b--', 
         xp, dgp, 'orange', 
         xn, gxn, 'b--',  
         xn, dgn, 'orange')
plt.legend(["y=g(x)" ,"y=g'(x)"])
plt.grid('on')
plt.xlim([-1,1])
plt.ylim([-2,2])
plt.show()
```

```python
def H(x):
    return x**2*np.sin(1/x**2)

h=1e-6
xp= np.linspace(1e-5,1)
xn= np.linspace(-1,-1e-5)
Hxp= H(xp)
Hxn= H(xn)
Hx1p = H(xp+h)
Hx1n = H(xn+h)
dHp = (Hx1p-Hxp)/h
dHn = (Hx1n-Hxn)/h

plt.plot(xp, dHp, 'orange', 
         xp, Hxp, 'b--',
         xn, dHn, 'orange',
         xn, Hxn, 'b--'  )
plt.legend(["y=H'(x)" ,"y=H(x)"])
plt.grid('on')
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.show()
```
