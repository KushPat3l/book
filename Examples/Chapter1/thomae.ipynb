```python
import numpy as np
import matplotlib.pyplot as plt

xlist = []
ylist = []

for q in range(2,200):
    for p in range(1,q):
        if np.gcd(p,q) == 1:
            xlist.append(p/q)
            ylist.append(1/q)

plt.plot(xlist, ylist, 'or', 
         markersize=3)
plt.xlim(0,1) 
plt.show()

lim =0.1
num = sum(y > lim for y in ylist)
print("Found %d points above %f" 
      %(num,lim))
```
