```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi_10 = round(np.pi,5)
print(pi_10)

x=np.arange(1.0,1000000.0)

S = np.zeros_like(x)
err = np.zeros_like(x)

old = np.zeros_like(x)
old[0] = np.sqrt(2)

S[0] = 2*(2/np.sqrt(2))

err[0] = abs(S[0] - np.pi)

for i in range(1,len(x)):
    old[i] = np.sqrt(2 + old[i-1])
    S[i] = S[i-1] * (2/old[i])
    err[i] = abs(S[i]-np.pi)

for i in range(0,len(err)):
    S[i]=round(S[i],5)
    
print(S[980000])

p=0
while S[p] != pi_10:
    p+=1
    if p == len(err):
        break
    
print('Number of iterations needed are ', p-1)
```

```python
err=np.zeros(p)
err[0] = abs(S[0] - pi_10)
err[1] = abs(S[1] - pi_10)
q=np.zeros_like(err)

for i in range(0, len(err)-1):
    err[i]=abs(S[i] - pi_10)
    err[i+1]=abs(S[i+1] - pi_10)
    q[i] = (np.log(err[i+1]/err[i]))/(np.log(err[i]/err[i-1]))
    #print(q[i])

Nplt = range(0, len(err)-2)  
plt.plot(Nplt, q[0:len(err)-2])
plt.show()
```
