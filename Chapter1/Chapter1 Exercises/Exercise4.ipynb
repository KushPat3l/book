```python
import numpy as np
import matplotlib.pyplot as plt

Nend = 100

x = np.arange(0,Nend+1)
F=np.zeros(Nend+1)
R=np.zeros(Nend+1)
Sum=np.zeros(Nend+1)

F[1] = 1.0
F[2] = 1.0
R[1] = 1.0
R[2] = 1.0
Sum[0] = 0.0
Sum[1] = 1.0
Sum[2] = 2.0

for i in np.arange(3, Nend+1):
    F[i] = F[i-1]+ F[i-2]
    R[i] = 1/F[i]
    Sum[i] = R[i] + Sum[i-1]

plt.plot(x,Sum)

p=3
while round(Sum[p-1],10) != round(Sum[p],10):
    p+=1
    if p == len(Sum):
        break

print(round(Sum[p-1],10))
print('The reciprocal Fibonacci constant to 10 decimal places is ', np.round(Sum[p],10))
```
