```python
import numpy as np
import matplotlib.pyplot as plt
import math

x=np.arange(1.0,1000.0)

S = np.zeros_like(x)
err = np.zeros_like(x)

def dfunc(k):
    return (((math.factorial(4*k))*(1103+26390*k))/(((math.factorial(k))**4)*((396)**(4*k))))

dlist = np.zeros_like(x)
dlist[0] = dfunc(0)
S[0] = 1/(((2*np.sqrt(2))/(9801))*dfunc(0))


for i in range(1,len(x)):
    dlist[i] = dlist[i-1] + dfunc(i)

err[0] = abs(S[0] - np.pi)

for i in range(1,len(x)):
    S[i] = 1/(((2*np.sqrt(2))/(9801))*dlist[i])
    err[i] = abs(S[i]-np.pi)
    
fig, (ax1, ax2) = plt.subplots(2,
                               figsize=(5,10))

ax1.plot(x,S,'.')
ax1.axhline(y=np.pi,color='r')
#ax1.set_ylim([3.0,3.7])
ax1.set_xlabel('Number of terms')
ax1.set_ylabel('Approximation')
ax1.grid('on')

ax2.plot(x,err)
#ax2.set_ylim([0,1])
ax2.set_xlabel('Number of terms')
ax2.set_ylabel('Error')
ax2.grid('on')
plt.show()
```
