```python
import numpy as np
import matplotlib.pyplot as plt

P = 5
Q = 5

Nend =1000

U = np.zeros(Nend+2)
R = np.zeros(Nend+2)

U[0] = 0.0
U[1] = 1.0
U[2] = P*U[1] - Q*U[0]

for i in np.arange(3,Nend):
    U[i]=P*U[i-1] - Q*U[i-2]
    R[i-1]=U[i]/U[i-1]

U[Nend]=P*U[Nend - 1] - Q*U[Nend - 2]  
U[Nend+1]=P*U[Nend] - Q*U[Nend - 1]
R[Nend-1]=U[Nend]/U[Nend-1]
R[Nend]=U[Nend+1]/U[Nend]

fig,(ax1,ax2)=plt.subplots(1,2,figsize=(11,4))

Nmin = 5
Nplt = range(Nmin, Nend+1)

ax1.semilogy(Nplt, U[Nmin:Nend+1], 'bo-', ms=3)
ax1.set_xlabel(r'$n$')
ax1.set_ylabel(r'$U_n$')
ax1.set_xlim(Nmin, Nend)
ax1.set_xticks(range(Nmin, Nend+1, 5))
ax1.grid()

ax2.plot(Nplt, R[Nmin:Nend+1], 'ro-', ms=3)
ax2.set_xlabel(r'$n$')
ax2.set_ylabel(r'$U_{n+1}/U_{n}$')
ax2.set_xlim(Nmin, Nend)
ax2.set_xticks(range(Nmin, Nend+1, 5))
ax2.grid()

plt.show()
```
