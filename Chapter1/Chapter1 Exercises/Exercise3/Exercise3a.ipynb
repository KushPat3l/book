```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(1.0,100.0)

S = np.zeros_like(x)
err = np.zeros_like(x)

def afunc(k):
    return ((-1)**(k+1))*(1/((2*k)-1))

S[0] = 4*afunc(1)

err[0] = abs(S[0] - np.pi)

for i in range(1,len(x)):
    S[i] = S[i-1] + (4*afunc(x[i]))
    err[i] = abs(S[i]-np.pi)
    
fig, (ax1, ax2) = plt.subplots(2,
                               figsize=(5,10))

ax1.plot(x,S, '.')
ax1.axhline(y=np.pi,color='r')
ax1.set_ylim([3.0,3.7])
ax1.set_xlabel('Number of terms')
ax1.set_ylabel('Approximation')
ax1.grid('on')

ax2.plot(x,err)
ax2.set_ylim([0,1])
ax2.set_xlabel('Number of terms')
ax2.set_ylabel('Error')
ax2.grid('on')
plt.show()
```
