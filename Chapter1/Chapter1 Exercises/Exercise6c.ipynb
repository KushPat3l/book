```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Used the code from Fibonacci.ipynb initially
Nend = 50

F = np.zeros(Nend+1)
R = np.zeros(Nend+2)

F[1] = 1
F[2] = 1
R[2] =1

for i in np.arange(3,Nend+1):
    F[i]=F[i-1]+F[i-2]
    R[i]=F[i]/F[i-1]
 
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(11,4))

Nmin = 5
Nplt = range(Nmin, Nend+1)

ax1.semilogy(Nplt, F[Nmin:Nend+1], 'bo-', ms=3)
ax1.set_xlabel(r'$n$')
ax1.set_ylabel(r'$F_n$')
ax1.set_xlim(Nmin, Nend)
ax1.set_xticks(range(Nmin, Nend+1, 5))
ax1.grid()

ax2.plot(Nplt, R[Nmin:Nend+1], 'ro-', ms=3)
ax2.set_xlabel(r'$n$')
ax2.set_ylabel(r'$F_{n}/F_{n-1}$')
ax2.set_xlim(Nmin, Nend)
ax2.set_xticks(range(Nmin, Nend+1, 5))
ax2.grid()

plt.show()
```

```python
#Main Bits
q = np.zeros(Nend+1)

for i in np.arange(3, Nend):
    q[i] = (np.log(R[i+1]))/(np.log(R[i]))

Nmin = 5
Nplt = range(Nmin, Nend)   

plt.plot(Nplt, q[Nmin:Nend], 'ro-', ms=3)
plt.xlabel(r'$n$')
plt.ylabel(r'$q$')
plt.xlim(Nmin, Nend)
plt.xticks(range(Nmin, Nend, 5))
plt.grid()

plt.show()
```
