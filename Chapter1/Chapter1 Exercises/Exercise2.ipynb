```python
import numpy as np
import matplotlib.pyplot as plt

#choose the number of books:
n = 40
books = float(n)

x = np.arange(1,books + 1)
overhang = np.zeros_like(x)
overhang[0]=0.0

for i in range(1,len(x)):
    overhang[i] = overhang[i-1] + (1/2)*(1/i)

plt.plot(x,overhang, '.')
plt.xlabel('Number of books')
plt.ylabel('Overhang (in units of books)')
plt.grid('on')
plt.show()
```

```python
#p is overhang wanted in cm
def want_overhang(p):
    want_unit = p/30.0
    k = 0
    while overhang[k] < want_unit:
        k+=1
    print('The number of books needed to create an overhang of', p, 'cm is', k)
    
want_overhang(31)
want_overhang(32)
want_overhang(50)
```
