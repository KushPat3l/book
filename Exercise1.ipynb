```python
import numpy as np
import matplotlib.pyplot as plt

epsilon = np.arange(1e-10, (1e-1) + (1e-13), 5e-4)
tracker = np.zeros_like(epsilon)
n = np.zeros_like(epsilon)
err = np.linspace(1,1,len(epsilon))

def error(k):
    return np.e - (1 + 1/k)**k

for k in range(0,len(epsilon)):
    while err[k] >= epsilon[k]:
        tracker[k] +=1
        err[k] = error(tracker[k])

#Verifying output
print(err)
print(tracker)
```

```python
n_min = np.zeros_like(epsilon)

for k in range(0,len(tracker)):
    n_min[k] = tracker[k] - 1
    
plt.plot(n_min, epsilon)
plt.ylim(1e-10,1e-1)
plt.xlabel('$N_{min}$')
epsilon = '\u03B5'
plt.ylabel(epsilon)
plt.semilogx(n_min)
plt.grid('on')
plt.show()
```
